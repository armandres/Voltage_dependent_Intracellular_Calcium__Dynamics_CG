# Estandarizacin de los datos

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

#Iniciar objeto StandardScaler
scaler = StandardScaler()
#columnas_a_normalizar
columnas_a_normalizar = df.iloc[:, 2:37]    
#Copia del dataframe original en donde se guardarn los datos normalizados
df_normalized = df.copy()
# Transformar los datos en las columnas seleccionadas
for columna in columnas_a_normalizar:
    df_normalized[columna] = scaler.fit_transform(df[[columna]])

#Para eliminar los NaN
#nueva_v_sin_nan = nueva_v.dropna()
print(df_normalized.head(10))
```

# Representacin bidimensional mediante tSNE (t-distributed Stochastic Neighbour Embedding)

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Inicializa el modelo t-SNE
tsne = TSNE(n_components=2, random_state=123)

# Reduce la dimensionalidad
X_tsne = tsne.fit_transform(df_normalized.iloc[:, 2:37])


# Se crea un DataFrame con los datos reducidos
df_tsne = pd.DataFrame(X_tsne, columns=['Componente 1', 'Componente 2'])

# Visualiza los datos t-SNE
plt.figure(figsize=(5, 5))
plt.scatter(df_tsne['Componente 1'], df_tsne['Componente 2'], alpha=0.5)
plt.title('Grfico t-SNE')
plt.xlabel('TSNE 1')
plt.ylabel('TSNE 2')
plt.grid(True)
plt.show()
```

```python
df_tsne.info()
```

```python
#import pandas as pd
df2 = df_normalized.join(df_tsne)
```

```python
#Se usa la funcin de pyplot para generar la leyenda del dendrograma
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as hc

#filas_temporales = df2.iloc[[59,86]]
#filtrado = df2.drop([59,86])

plt.figure(figsize=(18, 5))
dend = hc.dendrogram(hc.linkage(df2.iloc[:, 2:37], method = 'ward'))

#plt.scatter(dend, datos, cmap='viridis')
plt.title('Dendrograma')
plt.xlabel('Sperm (ID)')
plt.ylabel('Distancia Euclidiana')
plt.savefig("Dendrograma1.pdf", format='pdf')
plt.show()
```

```python
#Se usa la funcin de pyplot para generar la leyenda del dendrograma
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as hc

#filas_temporales = df2.iloc[[59,86]]
filtrado = df2.drop([59]) #59,86

plt.figure(figsize=(18, 5))
dend = hc.dendrogram(hc.linkage(filtrado.iloc[:, 2:37], method = 'ward'))

#plt.scatter(dend, datos, cmap='viridis')
plt.title('Dendrograma')
plt.xlabel('Sperm (ID)')
plt.ylabel('Distancia Euclidiana')
plt.savefig("Dendrograma2.pdf", format='pdf')
plt.show()
```

```python
#Aqu se importa el agrupamiento jerarquico desde la libreria scipy,
#esto nos ayudar a generar el dendograma
import scipy.cluster.hierarchy as hc

#Tambin se importa el algoritmo de agrupamiento aglomerativo desde sklearn
from sklearn.cluster import AgglomerativeClustering
#Ahora se usa el agrupamiento aglomerativo, se le pasa
#como parmetros el numero de grupos (de acuerdo al dendrograma)
#el mtodo de calculo de distancia y el criterio de conexin

clust = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')

filtrado = df2.drop([59])#valores atipicos de acuerdo del dendrograma
cluster = clust.fit_predict(filtrado.iloc[:, 2:37])
print(cluster)
print(len(cluster))
print(type(cluster))
print(cluster.shape)
```

```python
df_cluster = pd.DataFrame(cluster, columns=['cluster'])
df3 = filtrado.join(df_cluster)
#print(cluster.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener las columnas que quieres utilizar para colorear los puntos
columnas_de_coloreo = ["columna1", "columna2", "columna3"]  # Reemplaza con los nombres de tus columnas

# Definir el nmero de filas y columnas en la cuadrcula de subgrficos
num_filas = 2
num_columnas = (len(columnas_de_coloreo) + num_filas - 1) // num_filas

# Crear una figura y ejes para los subgrficos
fig, axs = plt.subplots(num_filas, num_columnas, figsize=(10, 8))

# Convertir los ejes en un array 1D si num_filas o num_columnas es igual a 1
if num_filas == 1:
    axs = axs.reshape(1, -1)
elif num_columnas == 1:
    axs = axs.reshape(-1, 1)

# Iterar sobre las columnas de coloreo y crear un scatterplot para cada una
for i, columna in enumerate(columnas_de_coloreo):
    fila = i // num_columnas
    columna = i % num_columnas
    ax = axs[fila, columna]
    sns.scatterplot(data=df3, x="Componente 1", y="Componente 2", hue=columna, ax=ax)
    ax.set(xlabel='tSNE 1', ylabel='tSNE 2', title=f"Coloreado por {columna}")

# Ajustar el diseo de los subgrficos
plt.tight_layout()
plt.show()
```
